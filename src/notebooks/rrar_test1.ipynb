# RRAR Board Test using PYNQ

This figure shows the basic function test configuration for RRAR Digital Board using PYNQ

```python
from rrar import rrar
```

## Create RRAR object

```python
a = rrar()
```

## Initialize RRAR H/W component

```python
a._setup_clock()
a._set_LED()
a._setup_afe()
a._init_jesdlink()
a._setup_pl_default()
```

## Rx1 port Capture & Plot , 10 times

```python
a._m_captplot(ch=1, domain='time', m_sample=16384, loop=10)
```

```python
a._m_captplot(ch=1, domain='freq', m_sample=16384, loop=10)
```
